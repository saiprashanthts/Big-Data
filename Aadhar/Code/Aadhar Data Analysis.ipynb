```
# Remove Duplicates
print(aadhar_df.count())
aadhar_df=aadhar_df.dropDuplicates()
# TOtally 9 duplicate records are present
print(aadhar_df.count())
# TOtally 2records are presnt which has more number of NA's
aadhar_df=aadhar_df.dropna(thresh=11)
print(aadhar_df.count())
```

```
# To Find the ratio of Number of Boys to Girls
df=aadhar_df.groupBy(['State','District','Gender']).agg({'*':'count'}).selectExpr('State','District','Gender','`count(1)` as Count').sort('Count',ascending=False)
df.createOrReplaceTempView('aadhar_district_tbl')
sqlContext.sql("select State,District,MaleCount,FemaleCount,MaleCount/FemaleCount as CountRatio from (select State,District,sum(case when Gender='M' then Count else 0 end) as MaleCount ,sum(case when Gender='F' then Count else 0 end) as FemaleCount  from aadhar_district_tbl group by State,District)tmp where MaleCount/FemaleCount != 0 and MaleCount/FemaleCount is not null order by CountRatio ").show(30)
```
